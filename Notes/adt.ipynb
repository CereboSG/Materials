```python
#YOUR_CODE_HERE
class Stack:

    def __init__(self,size):
        self.size = size
        self.array = [None]*self.size
        self.top = -1

    def push(self,obj):
        if not self.isFull():
            self.top += 1
            self.array[self.top]=obj
        else:
            print("Can't push. Stack is full.")

    def pop(self):
        if not self.isEmpty():
            popped = self.array[self.top]
            self.array[self.top]=None
            self.top -= 1
            return popped
        else:
            print('Nothing to pop. Stack is empty.')

    def peek(self):
        return self.array[self.top]

    def print(self):
        print(self.array)

    def isEmpty(self):
        return True if self.top == -1 else False

    def isFull(self):
        return True if self.top == self.size-1 else False


s=Stack(5)
s.print()

s.push(1)
s.print()
print(s.peek())

s.push(3)
s.print()
print(s.peek())

s.pop()
s.print()

new_s=Stack(5)
for i in range(5):
    new_s.push([j for j in range(i)])
    new_s.print()

new_s.push(6)

for i in range(len(new_s.array)-3):
    new_s.pop()

new_s.print()
```

```python
#YOUR_CODE_HERE
class Queue:
    
    def __init__(self,size):
        self.head = 0
        self.tail = -1
        self.size = size
        self.array = [None]*self.size

    def enqueue(self,obj):
        if not self.isFull():
            self.tail += 1
            self.array[self.tail]=obj
        else:
            print('Queue is full. Please wait until queue is no longer empty.')

    def dequeue(self):
        if not self.isEmpty():
            dequeued = self.array[self.head]
            self.array = self.array[1:]+[None]
            self.tail -= 1
            return dequeued
        else:
            print('Nothing to dequeue. Queue is empty.')
    
    def peek(self):
        return self.array[self.head]

    def print(self):
        print(self.array)

    def isFull(self):
        return True if self.tail == len(self.array) - 1 else False

    def isEmpty(self):
        return True if self.tail == -1 else False

q=Queue(3)
q.print()
q.enqueue(1)
q.enqueue(2)
q.print()
q.dequeue()
q.print()
q.peek()
q.enqueue(3)
q.enqueue(4)
q.print()
q.enqueue(5)
q.print()
q.dequeue()
q.print()
q.dequeue()
q.print()
q.dequeue()
q.print()
q.dequeue()
```
